```python
import pandas as pd
from scipy import stats
privacyInfo_df[['age_order','privacyLevel']].describe()
ttest,pval = stats.ttest_rel(privacyInfo_df['age_order'], privacyInfo_df['privacyLevel'])
print(pval)
if pval<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```

```python
import pandas as pd
from scipy import stats
from statsmodels.stats import weightstats as stests
ztest ,pval = stests.ztest(privacyInfo_df['privacyLevel'], x2=None, value=3)
print(float(pval))
if pval<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```

```python
import math
df = privacyInfo_df
df= df[:][df['pp_visit'] == 'Visited PP']
df['age_order'][df['age_order'] == 2] = 3
N = len(df)
age_s = 3

age2privacy = {1:[], 2:[], 3:[], 4:[], 5:[]} 
age_groups = range(age_s,9)
props_df = dftmp = pd.DataFrame([],columns=['proportion', 'age_order', 'privacyLevel'])
cumulative=False
omit_default = True
relative = True
for p in age2privacy.keys():
#     cntr = cntr +1
    if omit_default and p ==3:
        continue
    props = []; cum=0
    for j in age_groups:
       
        I = ((df['age_order'] == j) & (df['privacyLevel'] == p))
        if relative:
            N= sum((df['age_order'] == j))
        if cumulative:
            cum = cum + sum(I)/N
        else:
             cum = sum(I)/N       
        props.append(round(cum,4))
    dftmp = pd.DataFrame([],columns=['proportion', 'age_order', 'privacyLevel'])
    dftmp['proportion'] = props
    dftmp['age_order'] = age_groups
    dftmp['privacyLevel'] = p
    props_df = props_df.append(dftmp)
# print(props_df)

x,g = 'age_order', 'privacyLevel'

# (props_df
# .groupby(x)[g]
# .value_counts(normalize=True)
# .mul(100)
# .rename('percent')
# .reset_index()
# .pipe((sns.catplot,'data'), x=x,y='proportion',hue='privacyLevel',kind='point'))

sns.lineplot(data=props_df, x="age_order", y="proportion", hue="privacyLevel")

# x,y = 'privacyLevel', '__v'
# len_df = len(df.index)
# print(len_df)
# df = privacyInfo_df
# (df
# .groupby(x)[y]
# .value_counts(normalize=True)
# .mul(100)
# .rename('percent')
# .reset_index()
# .pipe((sns.catplot,'data'), x=x,y='percent',hue=y,kind='point'))
```
